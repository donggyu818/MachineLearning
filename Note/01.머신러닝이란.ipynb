## 머신러닝이란?
 : 수많은 데이터를 학습시켜 거기에 있는 패턴(규칙)을 찾아내고 이를 기반으로 데이터를 분류(Classification) 하거나 예측(predict).
  - 수식으로 얻을 수 없는 것들 (예_ 평균, 합계는 데이터가 아님)
  - 일반 소프트웨어(결정론적 소프트웨어)는 언제나 똑같은 과정을 거쳐 똑같은 결과를 냄.
  - 머신러닝 소프트웨어(비결정론적 소프트웨어)는 **데이터와 알고리즘**에 따라 때로는 다른 결과를 추론해냄. (데이터에 따라 알고리즘 중에 적절히 선택하면 됨)
    (추론 : 이미 알고있는 정보로부터 논리적 결론을 도출하는 과정)

#### 작업 방법
- 어느 정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석시킨다. (샘플 데이터는 원래 데이터의 성격을 가지고 있어야 함. 편향된 데이터를 뽑아선 안됨)
- 그리고 이때 데이터에서 일정한 규칙을 찾아 냅니다.
- 찾아낸 규칙을 기반으로 다른 데이터를 분류하거나 예측하는 것.

#### 어떻게 특징(Feature)과 규칙(Rule)을 찾을까?
- 특정 추출 데이터가 어떤 특징을 가지고 있는지 찾고 벡터(<=> 스칼라)로 만드는 처리과정. (벡터 : 데이터 형식이 같아야 함)
- 어떤 특징(Feature, Column)을 추출할지는 분석가가 결정. 이것이 머신러닝의 중요한 포인트.
- 회귀분석(y = ax + b. 분류와 달리 딱 뭐로 나누는 게 아니라 값을 주관식마냥 추출하는 거) : 데이터가 연속된 값일때 사용하며, 이를 선형회귀라고 부름.
    이때 y를 종속변수(Target), x를 독립변수(Feature) 라고 함.

#### 머신러닝의 종류

##### 지도학습 (Supervised Learning)
 : 학습 시 데이터와 답을 같이 학습한 후 다른 데이터의 답을 예측한다.

##### 비 지도학습 (Unsupervised Learning)
 : 학습 시 데이터만 학습한 후 다른 데이터의 규칙성을 찾는다.
  - 클러스터 분석(Cluster Analysis) => 분류
  - 주성분 분석(Pricipal Component Analysis : PCA)

##### 강화학습(Reinforcement Learning)
 - 부분적으로 답을 입력한 후 데이터를 기반으로 최적의 답을 찾아 낸다.

#### 머신러닝의 흐름

 (1) 데이터 수집 </br>
 (2) 데이터 가공 - 결측치, 이상치 관리 -> 분석. 숫자형으로 변환해야 함 </br>
 (3) 데이터 학습</br>
  (3.1) 학습 방법 선택</br>
  (3.2) 매개변수(Hyper Parameter) 조정</br>
  (3.3) 모델 학습</br>
 (4) 모델 평가</br>
 (5) 예측력과 오분류표가 좋은 경우 다음 단계를 진행하고 아니면 (3)을 재실행</br>
    - 예측력 : 정확도. 오분류표 : 개별 변수 별 정확도.(전체: 90%, 남:95%, 여:20% => 예측력은 좋으나, 오분류표가 안 좋음)</br>
 (6) 성공

#### 머신러닝의 응용 분야
##### 클래스 분류(Classification)
- 특정 데이터에 레이블(Target)을 분류할 수 있다.
- 스팸메일 분류, 필기인식, 증권사기
##### 이진분류
 : 둘 중 하나의 값으로 분류하는 경우
##### 다중분류
 : 여러 개의 분류 값 중에서 하나의 값으로 예측하는 경우

##### 클러스터링(Clustering) : 그룹 나누기
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있다.
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공.

##### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터를 추천하는 것.
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천.

##### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것.
- 데이터를 시각화하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약

#### 머신러닝의 필수 개념
##### 과대적합(Overfitting)
- 학습데이터 외의 데이터는 정확도가 낮게 나오는 모델. 학습데이터를 너무 공부해서 새 유형에 약한 거.

##### 과소적합(Underfitting)
- 학습데이터보다 일반데이터의 정확도가 높게 나오는 모델. 학습데이터 공부를 너무 안 해서 학습데이터에 약함.

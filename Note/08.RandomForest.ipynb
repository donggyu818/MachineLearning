### RandomForest
- 집단 학습을 기반으로 고정밀 분류, 회귀, 클러스터링 등을 구현
- 학습 전용 데이터를 기반으로 다수의 의사결정트리를 만들고 만들어진 의사결정 트리를 기반으로 다수결로 결과를 유도하는 모델

```python
import pandas as pd
```

```python
iris = pd.read_csv('../Data/iris.csv')
iris.head()
```

```python
# 이상이 없다면 Feature 와 Target 분리
iris_data = iris[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
iris_target = iris['Name']
```

```python
# Data 크기 확인
print(iris_data.shape)
print(iris_target.shape)
```

# Train 과 Test 분리

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_target, test_target = \
    train_test_split(
        iris_data,
        iris_target,
        random_state=42,
        stratify=iris_target
    )
```

```python
print(train_data.shape)
print(test_data.shape)
print(train_target.shape)
print(test_target.shape)
```

#### RandomForest 로 분류해보기

```python
from sklearn.ensemble import RandomForestClassifier  # Regressor : 주관식
```

```python
# 모델 만들기
clf = RandomForestClassifier()   # 기본값 : 100
```

```python
# 학습 시키기
clf.fit(train_data, train_target)
```

```python
# 평가하기
clf.score(test_data, test_target)
```

#### 랜덤포레스트를 이용한 버섯 분류

```python
import pandas as pd
```

```python
mr = pd.read_csv('../Data/mushroom.csv', header=None)
mr.head()   # 찍어봤더니 이상하게 나옴. => header = None
```

```python
mr.shape
```

- 한 줄이 버섯 한 종류
- 첫 번째 열 : p(독버섯), e(식용)
- 두 번째 열 : 버섯의 머리 모양 -> b(종 모양), c(원뿔 모양), x(볼록), f(평평), k(혹), s(오목)
- 네 번째 열 : 버섯의 머리 색깔 -> n(갈색), b(황갈색), c(연한갈색)
##### - Feature Data 가 알파벳 1자리 문자 -> Feature 는 숫자로 변환작업 해야함.

##### 연습

```python
# 알파벳 - ASCII(아스키) 코드
ord('x')
```

```python
# ASCII -> 알파벳
chr(120)
```

```python
# mr 의 1번 컬럼을 숫자로 변경. 하나씩 바꾸므로 for 문 사용함.
for i in mr.loc[:,1]:
    print(i, ord(i))
```

```python
data=[]
target = []

for row_index, row in mr.iterrows():        # iterrows : 한 줄씩 데이터 지정하는 거
    #print(row_index, ":", row)
    #print(row.loc[0])
    #print(row.loc[1])
    #print('----')
    target.append(row.loc[0])
    row_data = []        # [[], []] 식의 2차원 배열이기 때문에 data 안에 또 들어갈 row_data 를 만들어 줌
    for v in row.loc[1:]:
        row_data.append(ord(v))
    data.append(row_data)
```

```python
print(data[:2])
print(target[:2])
```

#### DataFrame 만들기

```python
# Target 을 기준으로 DataFrame 만들기
targetTemp = pd.DataFrame(target)
targetTemp.head()
```

```python
# data 를 DataFrame 으로 만들고 column 이름을 겹치지 않게 하기
dataTemp = pd.DataFrame(data)
dataTemp.rename(columns=lambda x: x+1, inplace=True)    # 람다함수 : 함수 이름이 없고 return 이 없음. 컬럼 번호를 +1 하게 함
dataTemp.head()
```

```python
# 2개의 dataframe 을 합치기
mr2 = pd.concat([targetTemp, dataTemp], axis='columns')   # columns 방향으로 합치기.
mr2.head()
```

```python
mr2.info()
```

#### RandomForest 로 분류하기

```python
mr2_data = mr2.loc[:, 1:]
mr2_data.head()
```

```python
mr2_target = mr2.loc[:,0]
mr2_target
```

```python
train_data_mr2, test_data_mr2, train_target_mr2, test_target_mr2 = \
    train_test_split(
        mr2_data,
        mr2_target,
        random_state=42,
        stratify=mr2_target
    )
```

```python
print(train_data_mr2.shape)
print(test_data_mr2.shape)
print(train_target_mr2.shape)
print(test_target_mr2.shape)
```

```python
clf.fit(train_data_mr2, train_target_mr2)
```

```python
print(clf.score(train_data_mr2, train_target_mr2))
print(clf.score(test_data_mr2, test_target_mr2))
```

#### One-hot Encoding
 - 숫자 데이터가 숫자로써 의미가 있으면 상관없지만, 위의 데이터는 알파벳으로 분류(빨강, 갈색)를 위한 데이터이므로 숫자 크기의 의미가 없다.
 - 이때 사용하는 것이 One-hot Encoding 이다.

##### 연습

```python
# 1번 열
dataTemp[1].unique()
```

```python
pd.get_dummies(
    data=dataTemp,
    columns=[1],
    prefix='1'   # 앞에 1이라고 적어줘라
)              # 1_??? 는 True 가 하나고, False 가 나머지이므로 이 행렬을 희소행렬이라 함.
```

```python
# 이미 만들어져서 에러 뜸
# 적용하기
dataTemp = mr2.copy()

for i in range(1, 23):
    dataTemp = pd.get_dummies(
        data=dataTemp,
        columns=[i],
        prefix=str(i)
    )
```

#### scikit learn 의 Machine Learning 종류

```python
import pandas as pd
```

```python
csv = pd.read_csv('../Data/iris.csv')
csv.head()
```

```python
# Feature 와 Target 나누기
iris_data = csv[[ 'SepalLength', 'SepalWidth',	'PetalLength',	'PetalWidth' ]]
iris_label = csv['Name']
```

```python
# Train 과 Test 구분
from sklearn.model_selection import train_test_split
```

```python
train_data , test_data, train_label, test_label = \
    train_test_split(
        iris_data,
        iris_label,
        random_state=42,
        stratify=iris_label
    )
```

주로 쓰는 모델만 알려줌. 모델 별로 점수를 출력하고, 점수가 젤 높은 모델을 쓰면 됨.
모델 각각 형식은 비슷하므로, 형태 이런 거 외우는 게 아니라 그냥 정해진 형식에 모델 불러와서 점수 보면 됨.

### kNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(train_data, train_label)
print("KNN :", knn_clf.score(test_data, test_label))
```

##### SVM

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(train_data, train_label)
print("SVM :", svm_clf.score(test_data, test_label))
```

#### Gaussian Process (가우시안)

```python
from sklearn.gaussian_process import GaussianProcessClassifier
gp_clf = GaussianProcessClassifier()
gp_clf.fit(train_data, train_label)
print("Gaussian :", gp_clf.score(test_data, test_label))
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(train_data, train_label)
print("Decision Tree :", dt_clf.score(test_data, test_label))
```

#### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(train_data, train_label)
print("RandomForest :", rf_clf.score(test_data, test_label))
```

#### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
ad_clf = AdaBoostClassifier()
ad_clf.fit(train_data, train_label)
print("AdaBoost :", ad_clf.score(test_data, test_label))
```

#### Neural Net

```python
from sklearn.neural_network import MLPClassifier
net_clf = MLPClassifier()
net_clf.fit(train_data, train_label)
print("Neural Net :", net_clf.score(test_data, test_label))
```

#### QDA

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda_clf = QuadraticDiscriminantAnalysis()
qda_clf.fit(train_data, train_label)
print("QDA :", qda_clf.score(test_data, test_label))
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(train_data, train_label)
print("Naive Bayes :", nb_clf.score(test_data, test_label))
```

```python
# Train 과 Test 로 분리 (위에서 data 와 target 이 분리되어 있으므로 바로 분리)
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = \
    train_test_split(
        iris_dataset['data'],
        iris_dataset['target'],
        random_state=42,
        stratify=iris_dataset['target']
    )
```

```python
# 확인 작업
print(train_input.shape)
print(train_target.shape)
print(test_input.shape)
print(test_target.shape)
```

#### kNN
 +) 주변 데이터 몇 개를 참조할 지 판단해봄

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)   # 주변 1개 가지고 판단해봄
```

```python
knn.fit(train_input, train_target)
```

```python
print(knn.score(train_input, train_target))
print(knn.score(test_input, test_target))
```

```python
import numpy as np    # 평균을 구하기 위한 np 임포트
```

```python
for i in range(1, 101, 2):   # kNN 이 짝수 개 나오면 안 되므로 2칸씩 띄워서 홀수로 나오게
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_input, train_target)
    pred = knn.predict(test_input)
    result = np.mean(pred == test_target)    # (?) 얘 의미가 pred == test_target 이 true 인 갯수 비율임?
    print(f"{i} : {result}")   # 3, 5, 9, 11 개를 썼을 때 최대임. 다 괜찮지만, 처리 속도 빠르게 하기 위해 최소치인 3개로.
```

#### 위를 함수로 만들기

```python
def kNNF(i):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_input, train_target)
    pred = knn.predict(test_input)
    result = np.mean(pred == test_target)
    print(f"{i} : {result}")
```

```python
for i in range(1, 11, 2):
    kNNF(i)
```
